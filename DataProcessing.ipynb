```python
import pandas as pd
```

```python
full_data = pd.read_csv('cannabis.csv')
full_data
```

```python
flavor_list =[]
for i, flavors in enumerate(full_data.get('Flavor')):
        flavorArr = str(flavors).split(',')
        for flavor in flavorArr:
            flavor_list.append([flavor, i])
            
flavor_list = pd.DataFrame(flavor_list, columns=['Flavor', 'Strain_ID'])
flavor_list.Flavor = flavor_list.Flavor.astype('string')
print("flavor list", flavor_list)

flavor_data = pd.Series(full_data['Flavor'].str.get_dummies(sep=',').columns, dtype='string')
print("flavor data", flavor_data)

flavor_xref = []
for index, flavor in flavor_data.items():
    
    for flav, strain_id in flavor_list.values:
        if flav == flavor:
            flavor_xref.append([index, strain_id])
            
flavor_xref = pd.DataFrame(flavor_xref, columns=['Flavor_id', 'Strain_id'])
print("flavor xref", flavor_xref)
```

```python
effect_list =[]
for i, effects in enumerate(full_data.get('Effects')):
        effectArr = str(effects).split(',')
        for effect in effectArr:
            effect_list.append([effect, i])
            
effect_list = pd.DataFrame(effect_list, columns=['Effects', 'Strain_ID'])
effect_list.Effects = effect_list.Effects.astype('string')
print("effect list", effect_list)

effect_data = pd.Series(full_data['Effects'].str.get_dummies(sep=',').columns, dtype='string')
print("effect data", effect_data)

effect_xref = []
for index, effect in effect_data.items():
    
    for flav, strain_id in effect_list.values:
        if flav == effect:
            effect_xref.append([index, strain_id])
            
effect_xref = pd.DataFrame(effect_xref, columns=['Effect_id', 'Strain_id'])
print("effect xref", effect_xref)
```

```python
type_list =[]
for i, types in enumerate(full_data.get('Type')):
        typeArr = str(types).split(',')
        for type in typeArr:
            type_list.append([type, i])
            
type_list = pd.DataFrame(type_list, columns=['Type', 'Strain_ID'])
type_list.Type = type_list.Type.astype('string')
print("type list", type_list)

type_data = pd.Series(full_data['Type'].str.get_dummies(sep=',').columns, dtype='string')
print("type data", type_data)

type_xref = []
for index, type in type_data.items():
    
    for flav, strain_id in type_list.values:
        if flav == type:
            type_xref.append([index, strain_id])
            
type_xref = pd.DataFrame(type_xref, columns=['Type_id', 'Strain_id'])
print("type xref", type_xref)
```
